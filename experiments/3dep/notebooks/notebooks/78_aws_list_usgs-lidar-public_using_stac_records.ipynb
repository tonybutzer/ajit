# Lets look at the json in the STAC

```python
import requests
```

```python
lidar_stac = 'https://usgs-lidar-stac.s3-us-west-2.amazonaws.com/ept/catalog.json'
```

```python
resp = requests.get(lidar_stac).json()
resp.keys()
```

```python
resp['links']
```

```python

my_co_links = []
for lnk in resp['links']:
    if 'CO_' in lnk['href']:
        print(lnk['href'])
        my_co_links.append(lnk['href'])
```

```python
my_co_links[1]
```

```python
import geopandas as gpd

g_df = gpd.read_file(my_co_links[1])
```

```python
g_df
```

```python
g_dfs = []
for lnk in my_co_links:
    g_df = gpd.read_file(lnk)
    g_dfs.append(g_df)
    
```

```python
import pandas as pd
my_co_gdf = pd.concat(g_dfs)
```

```python
my_co_gdf
```

```python
import json
from shapely.geometry import shape, GeometryCollection

with open("garden.geojson") as f:
  features = json.load(f)["features"]

# NOTE: buffer(0) is a trick for fixing scenarios where polygons have overlapping coordinates 
garden_shapely = GeometryCollection([shape(feature["geometry"]).buffer(0) for feature in features])
```

```python
new_gdf = my_co_gdf.loc[my_co_gdf.intersects(garden_shapely)]
```

```python
new_gdf
```

```python
import pickle

file = open('elpaso_gdf.pickle', 'wb')
pickle.dump(new_gdf, file)
file.close()
```
