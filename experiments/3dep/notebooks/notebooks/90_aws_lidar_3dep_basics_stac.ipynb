```python
import requests
```

```python
stac_catalog_url = 'https://usgs-lidar-stac.s3-us-west-2.amazonaws.com/ept/catalog.json'
```

```python
r = requests.get('https://usgs-lidar-stac.s3-us-west-2.amazonaws.com/ept/catalog.json')
```

```python
dir(r)
```

```python
cat_links = r.json()['links']
```

```python
links_list=[]
for link in cat_links:
    print (link['href'])
    links_list.append(link['href'])
```

```python
import os
for ref in links_list:
    bn = os.path.basename(ref)
    if bn.startswith('SD'):
        print(bn)
```

```python
yankton_list = [ref for ref in links_list if 'Yankton' in ref]
```

```python
yankton_list
```

```python
r = requests.get(yankton_list[0])
```

```python
r.json()
```

```python
r.json()['geometry']
```

```python
bb = r.json()['bbox']
```

```python
bb
```

```python
c_lon = bb[0]
c_lat = bb[1]
```

```python
import folium
m = folium.Map(location=[c_lat, c_lon], zoom_start=7, tiles='CartoDB positron')
```

```python
m
```

```python
import json
```

```python
geom_dict = r.json()['geometry']
```

```python
geoj = json.dumps(geom_dict)
geo_j = folium.GeoJson(data=geoj,style_function=lambda x: {'fillColor': 'orange'})
folium.Popup('Yankton Stuff').add_to(geo_j)
geo_j.add_to(m)
m
```
