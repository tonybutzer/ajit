```python
import os
def _mkdir(directory):
    try:
        os.stat(directory)
    except:
        os.mkdir(directory)
```

```python
def _run_pdal(file_basename):

    filen=f'/lidar/CO_Eastern_ElPaso_2018/{file_basename}'
    cmd = f'pdal info --all {filen}'
    results = subprocess.run(cmd, stdout=subprocess.PIPE, shell=True).stdout.decode('utf-8')
    return results
```

```python
import json
def make_stac_records(list_of_laz):
    stac_path = '/home/ec2-user/usgs_lidar_stac'
    _mkdir(stac_path)
    cnt=0
    for file_item in list_of_laz:
        # print(file_item)
        stac_name = file_item.replace('.laz', '_stac.json')
        #print(stac_name)
        results = _run_pdal(file_item)
        stac_record = run_stac_creator(results)
        stac_fullname = f'{stac_path}/{stac_name}'
        with open(stac_fullname, 'w') as outfile:
            json.dump(stac_record, outfile, sort_keys=True,indent=2, separators=(',', ': '))
        cnt = cnt + 1
        if not cnt%10:
            tot = len(list_of_laz)
            msg = f'processing {cnt} of {tot} -- {stac_name}'
            print(msg)
        
        
        
```
